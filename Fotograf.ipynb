```python
import tkinter as tk
from tkinter import Message ,Text
import cv2,os
import shutil
import csv
import numpy as np
from PIL import Image, ImageTk
import pandas as pd
import datetime
import time
import tkinter.ttk as ttk
import tkinter.font as font
from subprocess import Popen
from tkinter import messagebox
import time
from tkinter import simpledialog

window1 = tk.Tk()
window1.title("Görüntü İşleme")
window1.geometry('600x300')
window1.configure(background='grey')
message1 = tk.Label(window1, text="Yüz Tanıma Sistemi"  ,fg="white", bg="black"  ,font=('arial', 25, 'bold underline')) 
message1.place(x=140, y=35)





def Karsılastır():
    os.system("python 03_face_recognizer_camera.py")
    print("slm")
    
   #time.sleep(5.0)

    
    
def Düzenle():
    try:
        print("Fonksiyonlar Deneniyor")
        os.system("python recognizer.py")
        print("recognizer fonksiyon Tamamlandı")
        os.system("python 01_prepare_data.py")
        print("01_prepare_data fonksiyon Tamamlandı")
    except IOError:
        print("hata oluştu")
    finally:
        print("Fonksiyon Tamamlandı")
        
        

    
def DispWin():
    
    window = tk.Tk()
    window.title("Yüz Tanımlama")

    dialog_title = 'Çıkış'
    dialog_text = 'Emin misin?'
 
    window.geometry('1280x720')
    window.configure(background='black')


    window.grid_rowconfigure(0, weight=1)
    window.grid_columnconfigure(0, weight=1)


    message = tk.Label(window, text="Yüz Tanıma Kayıt Sistemi"  ,fg="white", bg="black"  ,font=('arial', 26, 'bold underline')) 

    message.place(x=420, y=50)

    lbl2 = tk.Label(window, text="İsminiz",width=20  ,bg="grey"    ,height=2 ,font=('arial', 15, ' bold ')) 
    lbl2.place(x=200, y=300)

    txt2 = tk.Entry(window,width=20  ,bg="white"  ,font=('arial', 15, ' bold ')  )
    txt2.place(x=600, y=315)
    def Video():
        name=(txt2.get())
        os.mkdir("data/Video/{name}".format(name=name))
        kamera =cv2.VideoCapture(0)
        fourcc=cv2.VideoWriter_fourcc(*'XVID')
        cıktı=cv2.VideoWriter("data/Video/"+name+"/{photo}_".format(photo=name)+ ".mp4",fourcc,10.0,(640,480))
        while(kamera.isOpened()):
            tf,video=kamera.read()
            if tf==True :
                video=cv2.flip(video,1)
                cıktı.write(video)
                cv2.imshow("Kaydet",video)
            if cv2.waitKey(100) & 0xFF == ord('q'):
                break
                
        kamera.release()
        cıktı.release()
        cv2.destroyAllWindows() 

    def clear():
        txt.delete(0, 'end')    
        res = ""
        message.configure(text= res)

    def clear2():
        txt2.delete(0, 'end')    
        res = ""
        message.configure(text= res)    
    
    def is_number(s):
        try:
            float(s)
            return True
        except ValueError:
            pass
 
        try:
            import unicodedata
            unicodedata.numeric(s)
            return True
        except (TypeError, ValueError):
            pass
 
        return False
 

 
    def FotografCek():        
        
        name=(txt2.get())
        os.mkdir("data/people/{name}".format(name=name))
        
        if(name != ""):
            cam = cv2.VideoCapture(0)
            harcascadePath = "haarcascade_frontalface_default.xml"
            detector=cv2.CascadeClassifier(cv2.data.haarcascades +harcascadePath)
            sampleNum=0
            while(True):
                ret, img = cam.read()
                img=cv2.flip(img,1)
                
                gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
                faces = detector.detectMultiScale(gray, 1.3, 5)
                for (x,y,w,h) in faces:
                    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)        
                    sampleNum=sampleNum+1
                    cv2.imwrite("data/people/"+name+"/{photo}_".format(photo=name)+ str(sampleNum) + ".jpg",img)
                    cv2.imshow('frame',img)
                if cv2.waitKey(100) & 0xFF == ord('q'):
                    break
                elif sampleNum>29:
                    break
            cam.release()
            cv2.destroyAllWindows() 
 
    
    clearButton2 = tk.Button(window, text="Temizle", command=clear2  ,width=5  ,height=1, activebackground = "Red" ,font=('times', 15, ' bold '))
    clearButton2.place(x=850, y=310)  
    
    VideoAl = tk.Button(window, text="VideoAl", command=Video  ,width=15  ,height=2, activebackground = "Red" ,font=('times', 17, ' bold '))
    VideoAl.place(x=50, y=500)
    
    takeImg = tk.Button(window, text="FotografCek", command=FotografCek  ,width=15  ,height=2, activebackground = "Red" ,font=('times', 17, ' bold '))
    takeImg.place(x=500, y=500)
    
    quitWindow = tk.Button(window, text="Çıkış", command=window.destroy  ,width=12  ,height=2, activebackground = "Red" ,font=('times', 17, ' bold '))
    quitWindow.place(x=950, y=500)
    window.mainloop()

Göster = tk.Button(window1, text="Göster", command=Karsılastır  ,width=10  ,height=2 ,activebackground = "Red" ,font=('times', 15, ' bold '))
Göster.place(x=50, y=140)
Düzenle = tk.Button(window1, text="Düzenle", command=Düzenle  ,width=10  ,height=2 ,activebackground = "Red" ,font=('times', 15, ' bold '))
Düzenle.place(x=250, y=140)
Kayıt = tk.Button(window1, text="Kayıt", command=DispWin  ,width=10  ,height=2 ,activebackground = "Red" ,font=('times', 15, ' bold '))
Kayıt.place(x=450, y=140)
message2 = tk.Label(window1, text="" ,bg="grey",activeforeground = "black",width=30  ,height=2  ,font=('arial', 15, ' bold ')) 
message2.place(x=180, y=210)

window1.mainloop()
```
